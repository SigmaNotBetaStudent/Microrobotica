## Compito
* Cercare un testo letterario ion formato .txt.
* Leggete in file.
* creare un dizionario che ha come chiavi tutti i caratteri presenti nel file .txt e come valore la frequenza assoluta del carattere nel testo.
* Creare un dizionario che ha come chiavi tutti i caratteri presenti nel file .txt e come valore delle liste. La lista contiene tutti i caratteri successori della lettera usata come chiave. Va bene che la lista contenga doppioni.

```python
# carico il testo dal file
with open('IlBaroneRampante.txt', 'r', encoding= 'utf-8') as file:
    text = file.read()

#creo un dizionario per le frequenze assolute dei dizionari
freq_char = {}
for char in text:
    freq_char[char] = freq_char.get(char, 0) + 1

#dizionario con i successori
nexts = {}
for i in range(len(text) - 1):
    char = text[i]
    next_char = text[i+1]

    if char not in nexts:
        nexts[char] = []
    nexts[char].append(next_char)


#stampa
print('Freq caratteri: ', freq_char)
print('Nexts: ', nexts)
    
```

# COMPITO: 
* Chiedere all'utente una lettera iniziale e la lunghezza del testo.
* Cercare nel dizionario la lettera iniziale e sorteggiare un carattere a caso dalla lista corrispondente
* Usare questo carattere come carattere successivo alla lettera iniziale
* A questo punto, la nuova lettera iniziale corrisponde al carattere appena sorteggiato e itero il procedimento fino a ottenere un testo della lunghezza desiderata
* Facoltativo: farlo con una sequenza di n lettere

```python
import random

# --- GENERAZIONE TESTO ---

# input utente
start_letter = input("Inserisci la lettera iniziale: ")
length = int(input("Inserisci la lunghezza del testo da generare: "))

# se la lettera iniziale non esiste nel dizionario, scegli a caso
if start_letter not in nexts:
    print("La lettera non  presente nel dizionario, ne scelgo una a caso.")
    start_letter = random.choice(list(nexts.keys()))

# inizializzazione della parola
word = start_letter

# ciclo di generazione
while len(word) < length:
    last = word[-1]
    if last not in nexts:  # se non ha successori, riparto da una lettera casuale
        last = random.choice(list(nexts.keys()))
    nextChar = random.choice(nexts[last])
    word += nextChar
    print("carattere:", nextChar)

print("\nTesto generato:", word)
```

## In Python tutto  un oggetto, anche le funzioni


```python
somma = lambda a, b: a+b
prodotto = lambda a,b: a*b
operazioni = {"+": somma, "*": prodotto} #non somma() e prodotto(), perch quando metto le tonde sto facendo una chiamata alla funzione
```

```python
operazioni
```

```python
op = input("Inserisci + per la somma * per il prodotto")
a = float(input("Inserisci il primo numero: "))
b = float(input("Inserisci il secondo numero"))

if op in operazioni:
    # una funzione
    print(operazioni[op](a, b))
else:
    print("Operazione non valida")
```
