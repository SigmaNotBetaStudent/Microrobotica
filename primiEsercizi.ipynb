Es1: Scrivi un programma che legga un numero da input e stampi se è pari o dispari.

```python
n= int(input('Inserisci numero: '))
print('Numero pari' if n%2==0 else 'Numero dispari')
```

Es2: Crea una list comprehension che, dato l’elenco nums = [1, 2, 3, 4, 5], produca la lista dei quadrati
dei numeri dispari.

```python
nums = [1, 2, 3, 4, 5]
quadrati_dispari = [n ** 2 for n in nums if n%2!=0]
print(quadrati_dispari)
```

Es4: Scrivi una funzione fattoriale(n) che calcoli il fattoriale di un numero in maniera ricorsiva.

```python
def fattoriale(n):
    if n<=1: #se non si mette la chiamata che ferma la funzione fattoriale, 
        return 1
    return n*fattoriale(n-1)

print(fattoriale(1558))
```
